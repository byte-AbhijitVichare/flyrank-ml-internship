## 2. Target or proxy

## Target or Proxy

The target is a proxy variable named **needs_refresh**.

A page is labeled as needing refresh if its impressions during the last 30 days are significantly lower than during the previous 30 days.

This target is derived from historical search performance rather than future information, making it suitable for supervised machine learning.

The output supports content teams in identifying pages that may require updates before performance declines further.

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

df["needs_refresh"] = (
    df["impressions_last_30d"] <
    0.8 * df["impressions_prev_30d"]
).astype(int)

print(df[["content_id","needs_refresh"]].head())
```

## 3. Success metric

## Success Metric

The primary evaluation metric will be the **F1-score**.

Since both declining and stable pages exist in the dataset, F1-score provides a balanced evaluation by considering both precision and recall.

A higher F1-score means the model correctly identifies pages needing refresh while reducing unnecessary recommendations.

The ML model will also be compared against a simple baseline classifier to ensure it provides meaningful improvements.

```python
print("Pages needing refresh:", df["needs_refresh"].sum())
print("Stable pages:", len(df) - df["needs_refresh"].sum())

print("\nPercentage needing refresh:")
print(round(df["needs_refresh"].mean()*100,2),"%")
```

## 5. Why ML beats a fixed rule here

## Why ML Beats a Fixed Rule

A simple rule such as "refresh every page older than 180 days" ignores many important signals including CTR, search position, engagement rate, content type, content age, and historical search performance.

Different pages behave differently even when they receive similar traffic.

A machine learning model can learn patterns across many features simultaneously and recommend the pages most likely to benefit from a refresh.

This produces more accurate and useful recommendations than relying on one fixed threshold.
