## 1. Build the feature vector

## Feature Vector

The feature vector contains historical search performance signals that are available before making a content refresh decision.

The features selected for this model are:

- impressions_last_30d
- clicks_last_30d
- ctr
- avg_position
- engagement_rate
- content_age_days
- word_count
- search_volume

These features describe how a page has performed historically without using future information.

Missing numerical values are filled with the median value of each feature before modeling.

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

feature_cols = [
    "impressions_last_30d",
    "clicks_last_30d",
    "ctr",
    "avg_position",
    "engagement_rate",
    "content_age_days",
    "word_count",
    "search_volume"
]

features = df[feature_cols].copy()

features = features.fillna(features.median(numeric_only=True))

print("Feature Vector Shape:", features.shape)
features.head()
```

## 3. The leakage hunt

## Leakage Check

Potential leakage fields were inspected before modeling.

Fields such as trend_direction and trend_pct summarize future performance and therefore should not be included in the feature vector.

Only historical information available before the refresh decision is retained.

```python
leakage_cols = [
    "trend_direction",
    "trend_pct"
]

print("Potential Leakage Columns")
print(df[leakage_cols].head())

safe_features = features.copy()

print("Safe Feature Columns")
print(safe_features.columns.tolist())
```

## 4. What I excluded and why

## Excluded Features

### trend_direction
Represents future movement and could leak label information.

### trend_pct
Calculated using future observations and therefore excluded.

### content_id
Acts only as an identifier and has no predictive value.

### client_id
Identifier only and may reduce model generalization.

Only historical performance features available at prediction time are included in the final feature vector.

```python
excluded = [
    "trend_direction",
    "trend_pct",
    "content_id",
    "client_id"
]

print("Excluded Features")
print(excluded)
```
