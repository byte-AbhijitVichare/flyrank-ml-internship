## 2. The question: decision, action, cost of a wrong call

## Research Question

### Decision

Which content pages should be reviewed first before search performance declines?

### Action

SEO or content teams can review the highest-priority pages and decide whether they should be refreshed, expanded, monitored, or left unchanged.

### Cost of a Wrong Recommendation

A false positive means time is spent reviewing a page that did not require attention.

A false negative means a genuinely declining page is ignored, potentially leading to reduced search visibility and traffic.

### Why ML?

Machine learning can combine multiple observable signals, such as impressions, CTR, average position, freshness, and query diversity, to rank pages for review more effectively than relying on a single rule.

```python
print("Decision: Prioritize content pages for review.")
print("Action: Refresh or monitor selected pages.")
print("Risk: False positives waste time; false negatives may lose traffic.")
```
